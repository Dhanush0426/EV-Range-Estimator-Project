```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('/content/electric_vehicles_spec_2025.csv.csv')
print("✅ Dataset Loaded Successfully!")
print(df.head())

# Check basic info
print("\nDataset Info:")
print(df.info())

# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Simple EDA – see relationships
plt.figure(figsize=(6,4))
sns.scatterplot(x='battery_capacity_kWh', y='range_km', data=df)
plt.title('Battery Capacity vs EV Range')
plt.show()

plt.figure(figsize=(6,4))
sns.scatterplot(x='efficiency_wh_per_km', y='range_km', data=df)
plt.title('Efficiency vs EV Range')
plt.show()
```
